```python
import numpy as np
import matplotlib.pyplot as plt
# %matplotlib ipympl
```

```python

t = np.array([])
yn = np.array([])
with open("dataset1.txt", "r") as file:
    for line in file:
        splitline = line.split()
        if len(splitline) == 2:
            t = np.append(t,float(splitline[0]))
            yn = np.append(yn,float(splitline[1]))
            
        
```

```python
plt.figure()
plt.plot(t,yn,label = "yn = y(t)")
plt.title("noisy data")
plt.xlabel("time")
plt.ylabel("yn")
plt.legend()
plt.show()
```

```python
# do lstsq

# Use column_stack to put the vectors side by side
M = np.column_stack([t, np.ones(len(t))])
# Use the lstsq function to solve for p_1 and p_2
(p1, p2), _, _, _ = np.linalg.lstsq(M, yn, rcond=None) # the output is a tuple
print(f"The estimated equation is {p1} t + {p2}")
```

```python
# plot estimated and original curve and compare
def stline(x, m, c):
    return m * x + c

y_est = stline(t, p1, p2)

plt.figure() # this is super cool!
plt.plot(t,yn,label = "noisy")
plt.plot(t,y_est, label = "estimated straight line")
plt.legend()
plt.show()
```
